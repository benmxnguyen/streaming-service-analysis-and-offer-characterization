### Clean: Types

```python
#capitalize values
merged_df['Type_x_x'] = merged_df['Type_x_x'].str.title()
#finding unique categorical values
print(f"Unique categorical values are {merged_df['Type_x_x'].append(merged_df['Type_y_x']).append(merged_df['Type_x_y']).append(merged_df['Type_y_y']).unique()}")

#standardizing values
merged_df['Type_x_y'].replace(0, 'Movie', inplace = True)
merged_df['Type_y_y'].replace(1, 'TV Show', inplace = True)
merged_df['Type_x_x'].replace(['Series', 'Episode'], ['TV Show','TV Show'], inplace = True)
```

```python
# populate relevant column, and if no values then drop
for i in range(len(merged_df['Type_x_x'])):
    if (pd.isnull(merged_df['Type_x_x'][i])):
        if (pd.isnull(merged_df['Type_y_x'][i])):
            if (pd.isnull(merged_df['Type_x_y'][i])):
                if (pd.isnull(merged_df['Type_y_y'][i])):
                    merged_df = merged_df.drop(i, axis = 0)
                else:
                    merged_df.loc[i,'Type_x_x'] = merged_df.loc[i,'Type_y_y']
            else:
                merged_df.loc[i,'Type_x_x'] = merged_df.loc[i,'Type_x_y']
        else:
            merged_df.loc[i,'Type_x_x'] = merged_df.loc[i,'Type_y_x']
```

```python
merged_df.reset_index(inplace = True)
merged_df.drop(['Type_y_x', 'Type_x_y', 'Type_y_y', 'index'], axis = 1, inplace = True)
merged_df.rename(columns = {'Type_x_x' : 'Offering Medium'}, inplace = True)
merged_df
```

### Clean: Rated/Rating

```python
# variables
anchor_column = 'Rated'
anchor_column_new_name = 'Advisory Rating'
collapse_column = 'Rating'

# list variable for tracking which rows have missing age ratings
dropped_list = []
# populate relevant column, and if no values then drop
for i in range(len(merged_df[anchor_column])):
    if (pd.isnull(merged_df[anchor_column][i])):
        if (pd.isnull(merged_df[collapse_column][i])):
#             dropped_list.append(i)
            merged_df = merged_df.drop(i, axis = 0)
        else:
            merged_df.loc[i, anchor_column] = merged_df.loc[i, collapse_column]

# fixing columns
merged_df.reset_index(inplace = True)
merged_df.drop([collapse_column, 'index'], axis = 1, inplace = True)
merged_df.rename(columns = {anchor_column : anchor_column_new_name}, inplace = True)
print(f"Dropping {len(dropped_list)}.")
merged_df
```
